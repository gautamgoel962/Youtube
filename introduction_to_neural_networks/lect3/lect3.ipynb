```python
def f(x): 
    return 3*x**2 - 4*x + 5
```

```python
f(3.0)
```

```python
# slope/gradient/differentation 
# slope = f(x + h) - f(h) / h

# analytically
x = 3.0
h = 0.0000001
slope = (f(x+h) - f(x)) / h
```

```python
print(slope)
```

```python
# calculus 

slope = 6 * x - 4
slope = 6 * 3.0 - 4

print(slope)
```
